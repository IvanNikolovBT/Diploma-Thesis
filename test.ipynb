```python
import re

text = """
<НАСЛОВ>Временската Плажа на Сеќерите</НАСЛОВ>
<PESNA>
На плажата на минатото, каде што времето се разбира само со очиглед,
светлината на сеќерите се плива во брановите на спомените.



Тука, секој миг е исполнет со сенки на оние што некогаш беа.
</PESNA>
"""

# Extract title
title_match = re.search(r'<НАСЛОВ>\s*(.*?)\s*</НАСЛОВ>', text, re.DOTALL)
name_of_new_song = title_match.group(1).strip() if title_match else 'no_title_found'

# Extract full song content, including all line breaks
song_match = re.search(r'<PESNA>\s*(.*?)\s*</PESNA>', text, re.DOTALL)
song_content = song_match.group(1).strip() if song_match else 'no_song_found'

print("Title:", name_of_new_song)
print("Song:\n", song_content)
```
