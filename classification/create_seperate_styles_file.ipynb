```python
import pandas as pd 
import re
CSV_PATH="../api_styles_all_in_one_text.csv"
database=pd.read_csv(CSV_PATH)
random_seed=47
database
```

```python
rows = []
for _,row in database.iterrows():
    pattern = r"(.+?):\s*(.+)"
    pairs = re.findall(pattern, row['extracted_styles'])
    author=row['author']
    song_title=row['song_title']
    print(f"Total pairs found: {len(pairs)}")
 
    for key, value in pairs:
        value_clean = value.strip()
        is_true = value_clean.lower().startswith("да")
        
        present_value = "Да" if is_true else "Не"
        
        new_row = {
            'author': author,
            'song_title': song_title,
            'extracted_style': key,
            'present': present_value,
            'bool': is_true
        }
        rows.append(new_row)

df = pd.DataFrame(rows)
df.to_csv("extracted_api_styles_seperate_categories.csv", index=False)
print("✅ Saved to extracted_api_styles_seperate_categories.csv")
```
